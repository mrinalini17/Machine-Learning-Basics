## TYPES OF ML

Unsupervised: Unlabelled data
    
Reinforcement: intelligent agents take actions->maximise rewards

## TYPES OF SUPERVISED LEARNINGS

1. Classification: for discrete values->true/false, male/female

Decision trees, Random Forest, K nearest neighbour

2. Regression: for continuous values-> salary, age

Logistic, Polynomial, Support Vector Machines

## TYPES OF UNSUPERVISED LEARNINGS

1. Clustering: grouping similar data points 

K means, Hierarchial

2. Association: relationship btw data points
